```python
import yfinance as yf
import pandas as pd
import datetime as dt
from pandas.tseries.offsets import MonthEnd
```

```python
# Let's get ticker symbols for all S&P 100 stocks
tickers = pd.read_html('https://en.wikipedia.org/wiki/S%26P_100')[2] #[2] isiliye kyuki is website ka 3rd table humare kaam ka hai to index 2 rahega
```

```python
tickers
```

```python
tickers = tickers.Symbol.to_list()
```

```python
df = yf.download(tickers,start='2013-1-1')
```

```python
# sirf adjusted close humare kaam ka hai
prices = df['Adj Close']
```

```python
prices
```

```python
# let's transform date index
prices.index = pd.to_datetime(prices.index)
```

```python
monthly_returns = prices.pct_change().resample('M').agg(lambda x: (x+1).prod() - 1)
```

```python
monthly_returns
```

```python
formation = dt.datetime(2013,2,28)
```

```python
formation
```

```python
# let's try to find out prior month returns
return_prev = monthly_returns.loc[formation - MonthEnd(1)]
```

```python
return_prev
```

```python
return_prev.name = 'monthly_returns'
```

```python
# transform this series to a dataframe so that we can add columns
return_prev = pd.DataFrame(return_prev)
```

```python
return_prev
```

```python
# add a decile column
return_prev['decile'] = pd.qcut(return_prev.monthly_returns,10,labels=False)
```
